```python
df = df.drop(df[df['Insulin']==0].index)
```

```python
df = df.drop(df[df['Glucose']==0].index)
```

```python
X = df.loc[:, df.columns != 'Outcome']
y = df['Outcome']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
scaler = StandardScaler().fit(X_train)
X_train_s = scaler.transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
def model_prediction(model, param):
    grid_search = GridSearchCV(model, param, scoring='accuracy')
    grid_search.fit(X_train, y_train)
    best_est = grid_search.best_estimator_
    y_pred_train = best_est.predict(X_train_s)
    y_pred_test = best_est.predict(X_test_s)
    train_score = accuracy_score(y_train, y_pred_train)
    test_score = accuracy_score(y_test, y_pred_test)
    print(train_score)
    print(test_score)
```

```python
#GradientBoostingRegressor
param_gbr = {"learning_rate": (0.1, 0.5, 1), "n_estimators": (50, 100, 500), "max_depth": (1, 5, 8)}
gbr = GradientBoostingClassifier()
model_prediction(gbr, param_gbr)
```

```python
#RandomForestClassifier
param_rf = {"n_estimators": (10, 50, 300), "max_depth": (1,8,10)}
rf = RandomForestClassifier()
model_prediction(rf, param_rf)
```

```python
#SVC
param_svc = {"C": (5, 10, 100)}
svc = SVC(kernel='linear')
model_prediction(svc, param_svc)
```
